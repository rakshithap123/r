# Do you enjoy Reading ?

Percentage of Enjoyement

```python
Enjoyement_of_reading=df.groupby(['Do you enjoy the Reading']).size()
plt.figure(figsize=(10,5))
plt.title("Enjoyment of reading")
plt.pie(Enjoyement_of_reading,labels=Enjoyement_of_reading.index,autopct='%2.1f%%',shadow=True)
plt.show()
```

 Deparment wise students enjoyment in reading

```python
grouped=df.groupby('Select your Faculty')
reading_enjoyement=grouped['Do you enjoy the Reading'].value_counts().reset_index(name='count')
sns.catplot(x='Select your Faculty',y='count',hue='Do you enjoy the Reading',data=reading_enjoyement,kind='bar')
plt.show()
```

* Based on the information presented in the two graphs, we can conclude that the majority of students enjoy their reading, and among all branches,
science stream students appear to enjoy it the most

# Does Pandemic Affected your reading Habit?

Percentage wise

```python
Percentage_wise=df.groupby(['Does Covid Pandemic Affect on your reading']).size()
plt.figure(figsize=(10,5))
plt.title("Impact of Pandemic on Reading Habit ")
plt.pie(Percentage_wise,labels=Percentage_wise.index,autopct='%2.1f%%',shadow=True)
plt.show()
```

Department wise

```python
grouped=df.groupby('Select your Faculty')
reading_enjoyement=grouped['Does Covid Pandemic Affect on your reading'].value_counts().reset_index(name='count')
sns.catplot(x='Select your Faculty',y='count',hue='Does Covid Pandemic Affect on your reading',data=reading_enjoyement,kind='bar')
plt.show()
```

* The prevailing opinion among most students is that the pandemic has had a negative impact on their studies
